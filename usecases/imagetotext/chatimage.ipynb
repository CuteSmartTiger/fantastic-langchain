```python
from langchain_openai import ChatOpenAI
```

```python
from langchain_core.messages import HumanMessage,SystemMessage
```

```python
# open ../../../static/img/brand/wordmark.png as base64 str
import base64
from pathlib import Path

from IPython.display import HTML

img_path = Path("./test.jpg")
img_base64 = base64.b64encode(img_path.read_bytes()).decode("utf-8")

# display b64 image in notebook
HTML(f'<img src="data:image/png;base64,{img_base64}">')
```

```python
llm = ChatOpenAI(base_url="https://api.aiproxy.io/v1", temperature=0, model="gpt-4-1106-vision-preview")
```

```python

messages = [
    SystemMessage(content=""),
    HumanMessage(
        content=[
            {
                "type": "image_url",
                "image_url": {
                    # langchain logo
                    "url": f"data:image/png;base64,{img_base64}",  # noqa: E501
                },
            },
            {"type": "text", "text": ""},
        ]
    )
]
llm.invoke(messages)
```
